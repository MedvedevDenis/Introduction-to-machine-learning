```python
import pandas as pd
from sklearn.ensemble import GradientBoostingClassifier, RandomForestClassifier
from sklearn.model_selection import train_test_split
import numpy as np
from sklearn.metrics import log_loss
import matplotlib.pyplot as plt
```

```python
df =pd.read_csv("./gbm-data.csv")
```

```python
df.head()
```

```python
data = df.values
```

```python
y, X = data[:, 0], data[:, 1:]
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.8, random_state=241
)
```

```python
def sigmoid(y_pred):
    return 1 / (1 + np.exp(-y_pred))
```

```python
learning_rates = np.array([1, 0.5, 0.3, 0.2, 0.1])

results = dict()

for learning_rate in learning_rates:
    clf = GradientBoostingClassifier(
        n_estimators=250, learning_rate=learning_rate, verbose=False, random_state=241
    )
    clf.fit(X_train, y_train)
    
    train_log_loss = [log_loss(y_train, sigmoid(y_pred_train)) for y_pred_train in clf.staged_decision_function(X_train)]
    test_log_loss = [log_loss(y_test, sigmoid(y_pred_test)) for y_pred_test in clf.staged_decision_function(X_test)]
    
    min_loss_index = np.argmin(test_log_loss)
    min_loss = test_log_loss[min_loss_index]
    
    results[learning_rate] = (min_loss_index, min_loss)
    
    plt.figure()
    plt.plot(train_log_loss, label="train")
    plt.plot(test_log_loss, label="test")
    plt.legend()
    plt.show()
```
